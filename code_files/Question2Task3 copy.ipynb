```python
from UCI_Dataset.csv.MakeTest import X_testing, y_testing
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
# X_test  = X_test.reshape(X_test.shape[0], -1)

print("Flattened Training data shape: ", X_train.shape)
# print("Flattened Testing data shape: ", X_test.shape)
```

```python
X = pd.DataFrame(X_train)
y = pd.Series(y_train, dtype="category")
```

```python
tree = DecisionTree("RD", max_depth=3, save_plot=True, plot_name="my_tree")
tree.fit(X, y)
dot = tree.plot()
display(dot)

y_hat = tree.predict(X)

print("Accuracy: ", accuracy(y_hat, y))
for cls in y.unique():
    print("Precision: ", precision(y_hat, y, cls))
    print("Recall: ", recall(y_hat, y, cls))
```

```python
X_testing = X_testing.reshape(X_testing.shape[0], -1)
X_testing_df = pd.DataFrame(X_testing)  
y_test = pd.Series(y_testing, dtype="category")

y_hat = tree.predict(X_testing_df)

print("Accuracy: ", accuracy(y_hat, y_test))
for cls in y.unique():
    print("Precision: ", precision(y_hat, y_test, cls))
    print("Recall: ", recall(y_hat, y_test, cls))
```
